```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### all flows

```python
file_name="./build/all_flows.csv"
```

```python
data = pd.read_csv(file_name)
data.columns = data.columns.str.replace(' ', '')
data = data.sort_values(by='size')
data['cdf'] = (np.arange(len(data)) + 1) / len(data)

```

```python
q25=data['size'].quantile(0.25)
q50=data['size'].quantile(0.50)
q75=data['size'].quantile(0.75)
q80=data['size'].quantile(0.80)
q90=data['size'].quantile(0.90)
print("q25:"+str(q25))
print("q50:"+str(q50))
print("q75:"+str(q75))
print("q80:"+str(q80))
print("q90:"+str(q90))
```

```python
plt.plot(data['size'], data['cdf'], marker='o', linestyle='-')
plt.xscale('log')
plt.xlabel('Size')
plt.ylabel('CDF')
plt.title('Cumulative Distribution Function (CDF)')
plt.grid(True)
plt.show()
```
